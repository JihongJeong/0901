```
import numpy as np
import matplotlib.pyplot as plt
```

```
N = 10_000
inside = []
for _ in range(N):
  x = np.random.uniform(-1, 1)
  y = np.random.uniform(-1, 1)
  z = np.random.uniform(-1, 1)
  if np.sqrt(x**2 + y**2 + z**2)<1:
    inside.append((x, y, z))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter([n[0] for n in inside], [n[1] for n in inside], [n[2] for n in inside], marker='.', alpha=0.5)
plt.show()
```

```
pi_estimated = 8 * (3/4) * (len(inside)/float(N))
print(pi_estimated)
```
